```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the two GeoJSON files
geojson1 = gpd.read_file(r"https://raw.githubusercontent.com/sothearith-min/Hexagons-of-56-Regions-from-AR6/refs/heads/main/zone_AR6.geojson")
geojson2 = gpd.read_file(r"https://raw.githubusercontent.com/sothearith-min/Hexagons-of-56-Regions-from-AR6/refs/heads/main/line_grouping.geojson")

# Plot both GeoJSON files on the same plot
fig, ax = plt.subplots(figsize=(15, 10))
geojson1.plot(ax=ax, alpha=0.8, edgecolor='w', linewidth=.8, color = "#80b299")  # No fill color, only edges
geojson2.plot(ax=ax, edgecolor='black', linewidth=.8, color = "black")  # Display lines in black

# Add the 'label' values as text to each polygon in geojson1
for idx, row in geojson1.iterrows():
    centroid = row['geometry'].centroid
    ax.text(centroid.x, centroid.y, row['label'], fontsize=10, ha='center',fontfamily="cursive")

ax.set_axis_off()
plt.show()
```
